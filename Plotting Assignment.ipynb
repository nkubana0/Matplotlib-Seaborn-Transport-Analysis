```python
plt.figure(figsize=(8,6))
plt.hist(data['comb'], bins=10, color='orange', edgecolor='black')
plt.title('Histogram of Combustion')
plt.xlabel('Combustion')
plt.ylabel('Frequency')
plt.show()
```

```python
# Only numeric columns
numeric_data = data.select_dtypes(include=['float64', 'int64'])

#correlation matrix
corr_matrix = numeric_data.corr()

# Create the heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', vmin=-1, vmax=1, linewidths=0.5)

plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```

**Interpreting the Heatmap**

**Dark Red Areas:** Indicate strong negative correlation.

**Dark Blue Areas:** Indicate a strong positive correlation.

**Light Colored Areas:** Indicate weak or no correlation between features.

**CO2 emissions and GHG Score:** CO2 emissions (co2) and greenhouse gas score (ghgScore) would be negatively correlated, meaning that vehicles with higher CO2 emissions tend to have a lower GHG score.

**Fuel Efficiency:** Features like comb (combined MPG), city, and highway are likely positively correlated with ghgScore, meaning better fuel efficiency often corresponds to a higher GHG score.

**Engine Displacement and Cylinders:** displ and cylinders are typically negatively correlated with ghgScore, because larger engines tend to emit more greenhouse gases.

Dropping Columns
If the target is the ghgScore, the columns to consider dropping would be:

**ID:** This column is an identifier and has no meaningful correlation with the target.
